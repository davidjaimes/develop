```python
import numpy as np
import pandas as pd
```

```python
def compute(pv, rate, pay, add, days):
    '''
    Generator that yields successive period payment details
    '''
    i = 0
    while pv > 0:
        interest = round(pv * rate * days[i] / 100 / 365, 2)
        sub = min(pv, pay + add - interest)
        pv -= sub
        yield sub, interest, pv
        i += 1
```

```python
rate = 4.33
per = 72
pv = 21887.7
date = '20171026'
st = (22, 'D')
add = 50
dt = pd.date_range(date, periods=per, freq='MS').shift(*st)
dt = dt.insert(0, pd.to_datetime(date))
days = np.array(np.diff(dt), dtype='timedelta64[D]')
days = np.array(days, dtype=int)
pay = round(-np.pmt(rate/100/12, per, pv), 2)
pd.DataFrame(list(compute(pv, rate, pay, add, days)))
```
