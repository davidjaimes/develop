```python
import brindle as bd
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

Paul, Gina, Tom, and Kelly all bought the same car worth 35,000 US dollars and financed it for 72 months. Paul got an APR of 5.42, Gina a 4.33, Tom a 3.12, and Kelly a 2.89. How much additional money each month towards the principal must each person make to keep the total interest over the lifetime of the loan at about 3,225 US dollars?

```python
df1, sm1 = bd.amortization(5.42, 72, 35000, '20181025', add=450)
df2, sm2 = bd.amortization(4.33, 72, 35000, '20181025', add=253)
df3, sm3 = bd.amortization(3.12, 72, 35000, '20181025', add=40)
df4, sm4 = bd.amortization(2.89, 72, 35000, '20181025', add=0)
```

```python
pd.DataFrame([sm1, sm2, sm3, sm4])
```

Plot your findings.

```python
plt.style.use(bd.greyfox)

plt.plot(df1['Date'], df1['End'], label='5.42%, $450')
plt.plot(df2['Date'], df2['End'], label='4.33%, $253')
plt.plot(df3['Date'], df3['End'], label='3.12%, $40')
plt.plot(df4['Date'], df4['End'], label='2.89%, $0')

plt.legend(fontsize=12)
plt.xlabel('Time')
plt.ylabel('US Dollars')
```
